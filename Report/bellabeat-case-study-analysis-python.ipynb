# 5. Analyze And Share

**renaming the column**

```python
daily_activity = daily_activity.rename(columns = {'ActivityDate' : 'Date'})
```

```python
hourly_calories = hourly_calories.rename(columns = {'ActivityHour' : 'Time'})
```

```python
hourly_steps = hourly_steps.rename(columns = {'ActivityHour' : 'Time'})
```

```python
sleep_day = sleep_day.rename(columns = {'SleepDay' : 'Date'})
```

**changing the data type**

```python
daily_activity['Date'] = pd.to_datetime(daily_activity['Date'])

hourly_calories['Time'] = pd.to_datetime(hourly_calories['Time'])

hourly_steps['Time'] = pd.to_datetime(hourly_steps['Time'])

sleep_day['Date'] = pd.to_datetime(sleep_day['Date'])
```

**Creating a day of the week column**

```python
daily_activity['day_of_week'] = daily_activity['Date'].dt.day_name()

sleep_day['day_of_week'] = sleep_day['Date'].dt.day_name()
```

```python
#Setting the style for the chart
sns.set_style('darkgrid')
```

# 5.1 Analysing Total Steps and Calories

# 5.1.1 Relationship between the total number of steps and calories

```python
daily_activity.agg(
                   {'TotalSteps' : ['min','max','mean'],
                    'Calories'   : ['min','max','mean']
                   }
)
```

```python
sns.scatterplot (data=daily_activity,
                x = 'TotalSteps',
                y = 'Calories',
                alpha = 0.8,
             )
plt.xlabel('Total Steps',size = 10)
plt.xlabel('Total Calories',size = 10)
plt.show()
```

* There is a **positive relationship between the Total_steps and Calories**, which indicates that the **greater the number of step**s taken, the **more calories the user burns**

* As the user increases their number of steps, their will be impact on calories burned and keeping active is crucial for maintaining good health

# 5.1.2 Average of Total Steps by Day of the Week

```python
daily_activity.groupby(daily_activity["day_of_week"])[["TotalSteps"]].mean()
```

```python
plot_1 = daily_activity.groupby(daily_activity["day_of_week"],as_index = False).mean()

sns.barplot(data = plot_1,
            x = 'day_of_week',
            y = 'TotalSteps',
           )
plt.xlabel('Weekdays',size = 10)
plt.ylabel('Total Steps',size = 10)
plt.xticks(rotation = 30)
plt.show()
```

* **Average daily steps is around 7k to 8k**, which is **lower** than the recommended steps per day.

* We can see that the **most active day** in terms of steps taken is **Saturday**, **followed by Tuesday**. The **least active** day is **Sunday** - most likely a rest day for the users.




# 5.1.3 Average of Total Steps by hour of the day

```python
hourly_steps.groupby(hourly_steps["Time"].dt.hour)[["StepTotal"]].mean()
```

```python
hourly_steps['hour'] = hourly_steps['Time'].dt.hour
plot_2 = hourly_steps.groupby(hourly_steps["hour"],as_index = False).mean()

sns.barplot (data = plot_2,
             x = 'hour',
             y = 'StepTotal')
plt.xlabel('hour of the day',size = 10)
plt.ylabel('Total Steps',size = 10)
plt.show()
```

* From the above data the users tend to **start becoming more active between 6 and 8 am**. 

* The **peak** hours seem to occur **between 5 and 7 pm**,users are likely choosing to work out after working hours.

* **After 8pm** the **activing level** starting **decreases**.

# 5.2 Analysing The Categories Of Activities

# 5.2.1 Different categories of Activities

```python
daily_activity.agg({'SedentaryMinutes': ['min','max','mean'] ,
                    'LightlyActiveMinutes' : ['min','max','mean'],
                    'FairlyActiveMinutes' : ['min','max','mean'],
                    'VeryActiveMinutes' : ['min','max','mean']
                   }
)
```

```python
plot_5 = daily_activity.agg({'SedentaryMinutes': 'mean' ,
                    'LightlyActiveMinutes' : 'mean',
                    'FairlyActiveMinutes' : 'mean',
                    'VeryActiveMinutes' : 'mean'
                   }
)
keys = ['SedentaryMinutes','LightlyActiveMinutes','FairlyActiveMinutes','VeryActiveMinutes']
explode = [0.1, 0.0, 0.0 ,0.0]

plt.pie(plot_5 , labels = keys, autopct='%.0f%%',explode = explode )
plt.show()
```

* The **Department of Health and Human Services** recommends at least **30 minutes** of moderate daily **physical activity**

* Even though from the above graph by combining **FairlyActiveMinutes + VeryActiveMinutes** we **achieve** the required goal.But we spent** large amount** of time in **sedentary minutes** which may lead to health issues.

# 5.3 Analyzing the categories of sleep time

# 5.3.1 Different categories of sleep time 

```python
sleep_day.agg({ 'TotalTimeInBed' : ['min','max','mean','sum'] ,
                'TotalMinutesAsleep' : ['min','max','mean','sum']
              }
)
```

```python
plot_10 = sns.scatterplot(data =sleep_day, x = 'TotalTimeInBed', y ='TotalMinutesAsleep',alpha = 0.7)
plt.show()
```

* From the above plot we can see a **strong positive linear association** between the **total number of minutes asleep and the total time spent in bed**, which means that most users are usually in bed only when they are sleeping, and not much longer before or after that.

# 5.3.2 Average differement categories of minutes by day of week

```python
sleep_day['AwakeTimeInBed'] = sleep_day['TotalTimeInBed'] - sleep_day['TotalMinutesAsleep']

sleep_day.groupby(['day_of_week'])[['TotalTimeInBed','TotalMinutesAsleep','AwakeTimeInBed']].mean()
```

```python
weekday = sleep_day.groupby(sleep_day["day_of_week"]).mean()
time_bed = weekday[['TotalMinutesAsleep','TotalTimeInBed','AwakeTimeInBed']]

time_bed.plot.bar(xlabel='Day of the Week', ylabel='Average of Minutes', rot=0, figsize=(8,5))
plt.show()
```

* According to **human health and Services** above **400 minutes** is ideal **sleep**.

* On **all** the **days** through out the **week** the users are having required sleep  and there is no significant change ,hence the sleep is consistent.

* But users tend to **sleep more** on **weekends** **than** the **weekdays** ,this may be the users choose this as a **rest day**

* we can still see that, **at times**, they do **spend a larger amount of time in bed** without being asleep. This could be **related to the weekends** when many people choose to relax.
